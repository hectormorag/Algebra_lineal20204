Una discusión breve sobre el producto interior de Frobenius. El producto interior de Frobenius es una operación binaria entre matrices que toma dos matrices y nos regresa un número. Está denotada de esta manera:

$$\langle A, B\rangle=\operatorname{traza}\left({ }^{t} A B\right)$$



Un ejemplo: hagamos el producto interior de Frobenius entre las siguientes matrices,

$$\mathbf{A}=\left(\begin{array}{ccc}
2 & 0 & 6 \\
1 & -1 & 2
\end{array}\right), \quad \mathbf{B}=\left(\begin{array}{ccc}
8 & -3 & 2 \\
4 & 1 & -5
\end{array}\right)$$

Primero, transponemos $A$:



$$\left(\begin{array}{ccc}
2 & 0 & 6 \\
1 & -1 & 2
\end{array}\right)^{T}$$

Entonces, 

$$\left(\begin{array}{cc}
2 & 1 \\
0 & -1 \\
6 & 2
\end{array}\right)$$

Ahora multiplicamos la matriz $A^t$ con la matriz $B$ y sumamos los elementos de la diagonal: 

$$\begin{aligned}
\langle\mathbf{A}, \mathbf{B}\rangle_{\mathrm{F}} &=2 \cdot 8+0 \cdot(-3)+6 \cdot 2+1 \cdot 4+(-1) \cdot 1+2 \cdot(-5) \\
&=16+12+4-1-10 \\
&=21
\end{aligned}$$



Entonces, así se sacan los productos interiores de Frobenius. 

El producto interior de Frobenius, induce la $\textbf{norma de Frobenius}$:





$$\|\mathbf{A}\|_{F}=\sqrt{\langle\mathbf{A}, \mathbf{A}\rangle_{F}}$$

$$\left|a\right|=\sqrt{\left\langle a, a\right\rangle}$$

$$\left|b\right|=\sqrt{\left\langle b, b\right\rangle}$$



Entonces, obtenemos que $cos (\theta)=-2 / \sqrt{5}$. Por lo tanto, 

$$\theta=\cos ^{-1}\left(\frac{-2}{\sqrt{5}}\right)$$

4. Sea $V$ el espacio de todas las funciones reales continuas y diferenciables en el intervalo $[a, b] .$

Definamos la siguiente operación binaria:

$$
\langle f, g\rangle=\int_{a}^{b} f(t) g(t) d t+\int_{a}^{b} f^{\prime}(t) g^{\prime}(t) d t
$$

Demuestre que $\langle,\rangle$ es un producto punto de $V$.

Las primeras dos propiedades se siguen inmediatamente de las propiedades de integrales.

Para la tercera propiedad desarrollemos la expresión: 

$$\int_{a}^{b} f^{2}+\int_{a}^{b}\left(f^{\prime}\right)^{2} \geq 0$$

Ahora, cuándo esto es igual a $0$ ? Recordemos de cálculo que si una función continua es positiva en todo el intervalo, entonces su integral es positiva. Entonces, como $f^{2}$ es continua, significa que $f=0$ en todo el intervalo. Por lo tanto $f^{\prime}=0$ en todo el intervalo, por lo tanto se cumple la desigualdad.

Por lo tanto, la operación es un producto punto.

5. Calcule los coeficientes de Fourier de la función onda triangular:

$$f(x)=\left\{\begin{array}{ll}
\frac{\pi}{2}+x, & \text { if }-\pi \leq x \leq 0 \\
\frac{\pi}{2}-x, & \text { if } 0<x \leq \pi
\end{array}\right.$$

definida en el intervalo $[-\pi, \pi]$.

La constante $a_0$ se calcula así:

$$\begin{array}{l}
a_{0}=\frac{1}{\pi} \int_{-\pi}^{\pi} f(x) d x=\frac{1}{\pi}\left[\int_{-\pi}^{0}\left(\frac{\pi}{2}+x\right) d x+\int_{0}^{\pi}\left(\frac{\pi}{2}-x\right) d x\right] \\
=\frac{1}{\pi}\left[\left.\left(\frac{\pi}{2} x+\frac{x^{2}}{2}\right)\right|_{-\pi} ^{0}+\left.\left(\frac{\pi}{2} x-\frac{x^{2}}{2}\right)\right|_{0} ^{\pi}\right]=\frac{1}{\pi}\left[0-\left(-\frac{\pi}{2}+\frac{(-\pi)^{2}}{2}\right)\right. \\
\left.+\left(\frac{\pi}{2}-\frac{\pi}{2}\right)-0\right]=0
\end{array}$$

Luego, para calcular $a_n$:

$$\begin{array}{l}
a_{n}=\frac{1}{\pi} \int_{-\pi}^{\pi} f(x) \cos n x d x=\frac{1}{\pi}\left[\int_{-\pi}^{0}\left(\frac{\pi}{2}+x\right) \cos n x d x+\int_{0}^{\pi}\left(\frac{\pi}{2}-x\right) \cos n x d x\right] \\
=\frac{1}{\pi}\left[\int_{-\pi}^{0} \frac{\pi}{2} \cos n x d x+\int_{-\pi}^{0} x \cos n x d x+\int_{0}^{\pi} \frac{\pi}{2} \cos n x d x-\int_{0}^{\pi} x \cos n x d x\right]
\end{array}$$

Integrando por partes, tenemos que:

$$\int x \cos n x d x=\frac{x \sin n x}{n}-\int \frac{x \sin n x}{n} d x=\frac{x \sin n x}{n}+\frac{\cos n x}{n^{2}}$$

Luego, 

$$\begin{array}{l}
a_{n}=\frac{1}{\pi}\left[\left.\frac{\pi}{2}\left(\frac{\sin n x}{n}\right)\right|_{-\pi} ^{0}+\left.\left(\frac{x \sin n x}{n}+\frac{\cos n x}{n^{2}}\right)\right|_{-\pi} ^{0}+\left.\frac{\pi}{2}\left(\frac{\sin n x}{n}\right)\right|_{0} ^{\pi}\right. \\
\left.-\left.\left(\frac{x \sin n x}{n}+\frac{\cos n x}{n^{2}}\right)\right|_{0} ^{\pi}\right]
\end{array}$$

Evaluando: 

$$\begin{array}{l}
a_{n}=\frac{1}{\pi n^{2}}\left[\left.(\cos n x)\right|_{-\pi} ^{0}-\left.(\cos n x)\right|_{0} ^{\pi}\right]=\frac{1}{\pi n^{2}}[\cos 0-\cos (-\pi n)-\cos \pi n+\cos 0] \\
=\frac{2}{\pi n^{2}}[1-\cos \pi n]=\frac{2}{\pi n^{2}}\left[1-(-1)^{n}\right]
\end{array}$$

Por lo tanto, obtuvimos los valores de los coeficientes de Fourier. 

Las funciones impares tienen el coeficiente $b_n=0$

Las funciones pares tiene los coeficientes $a_0$ y $a_n$ iguales a cero.
