# Problemas de ortogonalidad, ecuaciones e hiperplanos.

1. Encuentre los valores $x,y$ y $z$ (si es que existen) tal que el conjunto $A$ sea una base ortogonal de $\mathbb{R}^{3}$.

$$A=\{(x, 2,2) ;(1, y,-1) ;(-4,-1, z)\}$$

Una base para un espacio vectorial es una base ortogonal si cada par de vectores que conforman la base son ortogonales; es decir su producto interior es $0$. Los elementos de una base ortogonal son $\textbf{linealmente independientes}$ (se puede demostrar) y como podemos ver la dimensión de $A$ es la misma que la dimensión de $\mathbb{R}^{3}$. Entonces en nuestro ejercicio basta probar que los elementos son ortogonales.

$$\begin{array}{l}
\langle(x, 2,2),(1, y,-1)\rangle=x+2 y-2=0 \\
\langle(x, 2,2),(-4,-1, z)\rangle=-4 x-2+2 z=0 \\
\langle(1, y,-1),(-4,-1, z)\rangle=-4-y-z=0
\end{array}$$

Entonces nos queda un sistema de ecuaciones:

$$\begin{array}{r}
x+2 y=2 \\
-4 x+2 z=2 \\
-y-z=4
\end{array}$$

Obtenemos la matriz aumentada:

$$\left[\begin{array}{rrr|r}
1 & 2 & 0 & 2 \\
-4 & 0 & 2 & 2 \\
0 & -1 & -1 & 4
\end{array}\right]$$

Sacamos la matriz escalonada, 

$$\left[\begin{array}{rrr|r}
1 & 2 & 0 & 2 \\
0 & -1 & -1 & 4 \\
0 & 0 & -6 & 42
\end{array}\right]$$

Lo pasamos al siguiente sistema de ecuaciones, 

$$\begin{array}{l}
x+2 y=2 \\
-y-z=4 \\
-6 z=42
\end{array}$$

Resolvemos y obtenemos el resultado:

$$\begin{array}{l}
x=-4 \\
y=3 \\
z=-7
\end{array}$$


2. Encuentre la base dual de la siguiente base del espacio $V=\mathbb{R}^{3}$: 

$$\mathcal{B}=\left\{\overrightarrow{e_{1}}-\overrightarrow{e_{2}}, \overrightarrow{e_{1}}+\overrightarrow{e_{2}}, \overrightarrow{e_{3}}\right\}$$.

Hay que darnos cuenta que la definición de base dual es que, dada la base 

$$\beta=\left\{v_{1}, \ldots, v_{n}\right\}$$

su base dual es la base,

$$\beta^{*}=\left\{f_{1}, \ldots, f_{n}\right\}$$

tal que, 

$$f_{i}\left(v_{j}\right)=\delta_{ij}$$.

Entonces, consideremos nuestra $\beta=\left\{\overrightarrow{e_{1}}-\overrightarrow{e_{2}}, \overrightarrow{e_{1}}+\overrightarrow{e_{2}}, \overrightarrow{e_{3}}\right\}$, nosotros queremos calcular $\beta^{*}$. Además, sabemos que las $f_i$ son funciones lineales. Entonces se cumple que (por la definición del producto interior de nuestro espacio), $f_{i}\left(x_{1}, x_{2}, x_{3}\right)=a x_{1}+b x_{2}+c x_{3}$. Como sabemos, si definimos el comportamiento de $f_i$ en términos de nuestras bases, determinamos completamente a la función. 

Entonces, calculando el caso $i=1$:

$$\begin{aligned}
f_{1}\left(e_{1}-e_{2}\right) &=a-b=1 \\
f_{1}\left(e_{1}+e_{2}\right) &=a+b=0 \\
f_{1}\left(e_{3}\right) &=c=0
\end{aligned}$$

Entonces $a=-b$ lo cual implica que $-2b=1$ por lo que $b=-\frac{1}{2}$ y $a=\frac{1}{2}$; mientras que $c=0$. Entonces, 

$$f_{1}\left(x_{1}, x_{2}, x_{3}\right)=\frac{1}{2} x_{1}-\frac{1}{2} x_{2}$$

Calculando el caso $i=2$;

$$\begin{aligned}
f_{2}\left(e_{1}-e_{2}\right) &=a-b=0 \\
f_{2}\left(e_{1}+e_{2}\right) &=a+b=1 \\
f_{2}\left(e_{3}\right) &=c=0
\end{aligned}$$

Entonces, 

$$f_{2}\left(x_{1}, x_{2}, x_{3}\right)=\frac{1}{2} x_{1}+\frac{1}{2} x_{2}$$

Calculando el caso $i=3$:

$$\begin{aligned}
f_{3}\left(e_{1}-e_{2}\right) &=a-b=0 \\
f_{3}\left(e_{1}+e_{2}\right) &=a+b=0 \\
f_{3}\left(e_{3}\right) &=c=1
\end{aligned}$$

Entonces, 

$$f_{3}\left(x_{1}, x_{2}, x_{3}\right)=x_3$$

Por lo tando la base dual es:

$$\beta^{*}=\left\{f_{1}, f_{2}, f_{3}\right\}$$





3. Sea $V=\mathbb{R}^{2} .$ Consideremos la base $B=\{(1,1),(1,-1)\} .$  Encuentre una base dual:


Si $B^{*}=\left\{\varphi_{1}, \varphi_{2}\right\}$ es la base dual de $B$, entonces debe cumplir


$$
\left\{\begin{array}{l}
\varphi_{1}(1,1)=1 \\
\varphi_{1}(1,-1)=0
\end{array} \quad y \quad\left\{\begin{array}{l}
\varphi_{2}(1,1)=0 \\
\varphi_{2}(1,-1)=1
\end{array}\right.\right.
$$

Puesto que para cada $(x, y) \in \mathbb{R}^{2}$ vale

$$
\begin{array}{c}
(x, y)=\frac{x+y}{2}(1,1)+\frac{x-y}{2}(1,-1) \\
\text { resulta que } \varphi_{1}(x, y)=\frac{x+y}{2} \text { y } \varphi_{2}(x, y)=\frac{x-y}{2}
\end{array}
$$

4. Sea $V=\mathbb{R}^{2} .$ Consideremos la base $B=\{(2,1),(3,1)\} .$  Encuentre una base dual:

Por definición, la base dual es el conjunto de todas las transformaciones tales que, 

$$\varphi_{i}\left(\mathbf{v}_{\mathbf{j}}\right)=\delta_{i j}$$

entonces, 

$$\begin{array}{l}
\varphi_{1}\left(\mathbf{v}_{1}\right)=\delta_{11}=1 \Leftrightarrow \varphi_{1}(2,1)=1 \Leftrightarrow \varphi_{1}[2(1,0)+1(0,1)]=1 \Leftrightarrow 2 \varphi_{1}(1,0)+1 \varphi_{1}(0,1)=1 \\
\varphi_{1}\left(\mathbf{v}_{2}\right)=\delta_{12}=0 \Leftrightarrow \varphi_{1}(3,1)=0 \Leftrightarrow \varphi_{1}[3(1,0)+1(0,1)]=0 \Leftrightarrow 3 \varphi_{1}(1,0)+1 \varphi_{1}(0,1)=0
\end{array}$$

Resolviendo el sistema:

$$\left[\begin{array}{ll}
2 & 1 \\
3 & 1
\end{array}\right]\left[\begin{array}{l}
\varphi_{1}(1,0) \\
\varphi_{1}(0,1)
\end{array}\right]=\left[\begin{array}{l}
1 \\
0
\end{array}\right]$$

Tenemos que, 

$$\varphi_{1}(1,0)=-1, \quad \varphi_{1}(0,1)=3$$

Entonces, 

$$\varphi_{1}(x, y)=x \varphi_{1}(1,0)+y \varphi_{1}(0,1)=-x+3 y$$

De forma similar: 

$$\varphi_{2}(x, y)=x \varphi_{2}(1,0)+y \varphi_{2}(0,1)=x-2 y$$

Por lo tanto, la base dual que estamos buscando será: $\left\{\varphi_{1}, \varphi_{2}\right\}=\{-x+3 y, x-2 y\}$.

Antes de terminar, una observación interesante: 

Supongamos que tenemos la función, 

$$\varphi=8 x-7 y$$

Resulta que la podemos escribir como combinación lineal de los elementos de la base y los elementos de la base dual: 

$$\varphi=\varphi\left(\mathbf{v}_{\mathbf{1}}\right) \varphi_{1}+\varphi\left(\mathbf{v}_{2}\right) \varphi_{2}+\ldots+\varphi\left(\mathbf{v}_{\mathbf{n}}\right) \varphi_{n}$$

De la siguiente forma: 

$$\begin{array}{l}
8 x-7 y=\overbrace{\varphi(2,1)}^{\varphi\left(\mathbf{v}_{1}\right)} \cdot \underbrace{(-x+3 y)}_{\varphi_{1}}+\overbrace{\varphi(3,1)}^{\varphi\left(\mathbf{v}_{2}\right)} \cdot \underbrace{(x-2 y)}_{\varphi_{2}} \\
8 x-7 y=(8 \cdot 2-7 \cdot 1) \cdot(-x+3 y)+(8 \cdot 3-7 \cdot 1) \cdot(x-2 y) \\
8 x-7 y=9(-x+3 y)+17(x-2 y) \\
8 x-7 y=8 x-7 y
\end{array}$$

Recordando una definición:

Una función $F: V \times V \rightarrow K$ es un producto interno o escalar si satisface las siguientes propiedades:

$$\langle\alpha \bar{u}+\beta \bar{v} \mid \bar{w}\rangle=\alpha\langle\bar{u} \mid \bar{w}\rangle+\beta\langle\bar{v} \mid \bar{w}\rangle \quad  \text{Linealidad}$$ 

$$\langle\bar{u} \mid \bar{v}\rangle=\langle\bar{v} \mid \bar{u}\rangle \quad \text{Simetría} $$


$$\langle\bar{u} \mid \bar{u}\rangle>0, \bar{u} \neq \overline{0} \quad \text{Positividad}$$

5. Verifique que el producto punto en $\mathbb{R}^{2}$ es un producto interior.

$$\left(x_{1}, y_{1}\right) \cdot\left(x_{2}, y_{2}\right)=x_{1} x_{2}+y_{1} y_{2}$$

Verificamos la linealidad: 

$$\begin{aligned}
\left[\alpha\left(x_{1}, y_{1}\right)+\beta\left(x_{2}, y_{2}\right)\right] \cdot\left(x_{3}, y_{3}\right) &=\alpha\left(x_{1}, y_{1}\right) \cdot\left(x_{3}, y_{3}\right)+\beta\left(x_{2}, y_{2}\right) \cdot\left(x_{3}, y_{3}\right) \\
\left(\alpha x_{1}+\beta x_{2}, \alpha y_{1}+\beta y_{2}\right) \cdot\left(x_{3}, y_{3}\right) &=\alpha\left(x_{1} x_{3}+y_{1} y_{3}\right)+\beta\left(x_{2} x_{3}+y_{2} y_{3}\right) \\
\left(\alpha x_{1}+\beta x_{2}\right) x_{3}+\left(\alpha y_{1}+\beta y_{2}\right) y_{3} &=\alpha x_{1} x_{3}+\alpha y_{1} y_{3}+\beta x_{2} x_{3}+\beta y_{2} y_{3} \\
\alpha x_{1} x_{3}+\beta x_{2} x_{3}+\alpha y_{1} y_{3}+\beta y_{2} y_{3} &=\alpha x_{1} x_{3}+\alpha y_{1} y_{3}+\beta x_{2} x_{3}+\beta y_{2} y_{3}
\end{aligned}$$

Verificamos la propiedad de simetría: 

$$\begin{aligned}
\left(x_{1}, y_{1}\right) \cdot\left(x_{2}, y_{2}\right) &=\left(x_{2}, y_{2}\right) \cdot\left(x_{1}, y_{1}\right) \\
x_{1} x_{2}+y_{1} y_{2} &=x_{2} x_{1}+y_{2} y_{1}
\end{aligned}$$

Verificamos la propiedad de que el producto interior siempre es positivo:

$$\begin{aligned}
\left(x_{1}, y_{1}\right) \cdot\left(x_{1}, y_{1}\right) &=x_{1} x_{1}+y_{1} y_{1} \\
&=x_{1}^{2}+y_{1}^{2}
\end{aligned}$$
    

Por lo tanto, sí es un producto interior.
